```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/STEAM/AiElements Data Science Internship Assignment Solution/Data/ModelData.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
y = df.price
X = df.drop('price', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

Linear Regression

```python
LinearRegression_Model = LinearRegression()
LinearRegression_Model.fit(X_train, y_train)
print(LinearRegression_Model.score(X_test, y_test)) #R2 score
```

Random Forest Regressor

```python
RandomForest_Model = RandomForestRegressor(n_estimators = 2000, random_state = 42)
RandomForest_Model.fit(X_train, y_train);
print(RandomForest_Model.score(X_test, y_test)) #R2 score
```

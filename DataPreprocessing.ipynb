```python
def To_number(x):
    return x.map({'two':2,'four':4, 'five': 5,'six': 6,'eight':8})
    
df['cylindernumber'] = df[['cylindernumber']].apply(To_number)
df["cylindernumber"] = df["cylindernumber"].astype(np.int64)
df.cylindernumber.unique()
```

```python
OneHotEncodingColomn = ['fueltype', 'aspiration','doornumber','carbody','drivewheel','enginelocation','Company','fuelsystem','enginetype'] 

final_df = pd.get_dummies(df, columns = OneHotEncodingColomn)
```
